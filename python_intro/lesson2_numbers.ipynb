### 2.1 –ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏

**–û—Å–Ω–æ–≤–Ω—ã–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏ Python**

| –û–ø–µ—Ä–∞—Ü–∏—è | –ù–∞–∑–≤–∞–Ω–∏–µ              | –ü—Ä–∏–º–µ—Ä   | –†–µ–∑—É–ª—å—Ç–∞—Ç |
| -------- | --------------------- | -------- | --------- |
| `+`      | —Å–ª–æ–∂–µ–Ω–∏–µ              | `5 + 2`  | `7`       |
| `-`      | –≤—ã—á–∏—Ç–∞–Ω–∏–µ             | `5 - 2`  | `3`       |
| `*`      | —É–º–Ω–æ–∂–µ–Ω–∏–µ             | `5 * 2`  | `10`      |
| `/`      | –¥–µ–ª–µ–Ω–∏–µ               | `5 / 2`  | `2.5`     |
| `//`     | —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–µ –¥–µ–ª–µ–Ω–∏–µ | `5 // 2` | `2`       |
| `%`      | –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è    | `5 % 2`  | `1`       |
| `**`     | –≤–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å  | `2 ** 3` | `8`       |

üí° –í—Å–µ —ç—Ç–∏ –æ–ø–µ—Ä–∞—Ü–∏–∏ –º–æ–∂–Ω–æ –¥–µ–ª–∞—Ç—å —Å –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏, —á–∏—Å–ª–∞–º–∏ –∏ –¥–∞–∂–µ —Å –≤—ã—Ä–∞–∂–µ–Ω–∏—è–º–∏ –≤–Ω—É—Ç—Ä–∏.




–ó–∞–¥–∞–¥–∏–º –¥–≤–∞ —á–∏—Å–ª–∞ `a` –∏ `b`:

````python
a = 100
b = 4
````

````python
# —Å–ª–æ–∂–µ–Ω–∏–µ: +
a + b
````

````python
# –≤—ã—á–∏—Ç–∞–Ω–∏–µ: -
a - b
````

````python
# —É–º–Ω–æ–∂–µ–Ω–∏–µ
a * b
````

````python
# –¥–µ–ª–µ–Ω–∏–µ - —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤—Å–µ–≥–¥–∞ float!
a / b
````

````python
# –¥–µ–ª–µ–Ω–∏–µ –Ω–∞—Ü–µ–ª–æ - —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤—Å–µ–≥–¥–∞ int
a // b
````

````python
# –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è
a % b
````

````python
# —Ç–æ–∂–µ –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç —Ü–µ–ª–æ–≥–æ –¥–µ–ª–µ–Ω–∏—è (15 % 4 = 15 - 4 * (15 // 4) = 15 - 4 * 3 = 15 - 12 = 3)
15 % 4
````

````python
# –≤–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å
2 ** 4
````

–û–±–æ–±—â–∏–º –≤ –æ–¥–Ω–æ–π —è—á–µ–π–∫–µ:
- –ó–∞–¥–∞–¥–∏–º –¥–≤–∞ —á–∏—Å–ª–∞ –∏ –ø–æ–ª–æ–∂–∏–º –∏—Ö –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ
- –í—ã–≤–µ–¥–µ–º –Ω–∞ —ç–∫—Ä–∞–Ω —ç—Ç–∏ –¥–≤–∞ —á–∏—Å–ª–∞
- –î–ª—è –∫–∞–∂–¥–æ–π –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–π –æ–ø–µ—Ä–∞—Ü–∏–∏ –±—É–¥–µ–º –ø–∏—Å–∞—Ç—å –µ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ —á–∏—Å–ª–∞–º–∏

````python
a = 10
b = 3

print("–ß–∏—Å–ª–∞:", a, b)
print("–°–ª–æ–∂–µ–Ω–∏–µ:", a + b)
print("–í—ã—á–∏—Ç–∞–Ω–∏–µ:", a - b)
print("–£–º–Ω–æ–∂–µ–Ω–∏–µ:", a * b)
print("–î–µ–ª–µ–Ω–∏–µ:", a / b)
print("–¶–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–µ –¥–µ–ª–µ–Ω–∏–µ:", a // b)
print("–û—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è:", a % b)
print("–í–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å:", a ** b)
````

### 2.2 –û–∫—Ä—É–≥–ª–µ–Ω–∏–µ

–£ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª –µ—Å—Ç—å "—Ö–≤–æ—Å—Ç" - –∏—Ö –¥–µ—Å—è—Ç–∏—á–Ω–∞—è —á–∞—Å—Ç—å. –ù–∞–ø—Ä–∏–º–µ—Ä, —É —á–∏—Å–ª–∞ `3.14159` –µ–≥–æ —Ö–≤–æ—Å—Ç - `14159`. –ú—ã –º–æ–∂–µ–º –æ–∫—Ä—É–≥–ª—è—Ç—å —Ç–∞–∫–∏–µ —á–∏—Å–ª–∞ —Ä–∞–∑–Ω—ã–º–∏ —Å–ø–æ—Å–æ–±–∞–º–∏. –í Python –µ—Å—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –≤–∏–¥–æ–≤ –æ–∫—Ä—É–≥–ª–µ–Ω–∏—è:
1. `int()` - –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –≤–Ω–∏–∑ (–æ—Ç–±—Ä–∞—Å—ã–≤–∞–µ—Ç –¥—Ä–æ–±–Ω—É—é —á–∞—Å—Ç—å)
2. `round()` - –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –ø–æ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º –ø—Ä–∞–≤–∏–ª–∞–º
3. `math.floor()` - –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –≤–Ω–∏–∑ (–¥–æ –±–ª–∏–∂–∞–π—à–µ–≥–æ —Ü–µ–ª–æ–≥–æ)
4. `math.ceil()` - –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –≤–≤–µ—Ä—Ö (–¥–æ –±–ª–∏–∂–∞–π—à–µ–≥–æ —Ü–µ–ª–æ–≥–æ)

–ó–∞–¥–∞–¥–∏–º —Ç—Ä–∏ —á–∏—Å–ª–∞ - —Å –Ω–∏–º–∏ –∏ –±—É–¥–µ–º —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å

````python
a = 2.49230
b = 2.5969534
c = 2.934234
````

#### **`int()`** - –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –≤–Ω–∏–∑
- –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Ç–∏–ø `float()` –≤ `int()`
- –û—Ç–±—Ä–∞—Å—ã–≤–∞–µ—Ç –¥—Ä–æ–±–Ω—É—é —á–∞—Å—Ç—å, —Ç–æ –µ—Å—Ç—å –æ–∫—Ä—É–≥–ª—è–µ—Ç **–≤–Ω–∏–∑** –¥–æ –±–ª–∏–∂–∞–π—à–µ–≥–æ —Ü–µ–ª–æ–≥–æ

````python
print(a, '->', int(a))
print(b, '->', int(b))
print(c, '->', int(c))
````

#### **`round()`** - –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –ø–æ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º –ø—Ä–∞–≤–∏–ª–∞–º
- –µ—Å–ª–∏ –¥—Ä–æ–±–Ω–∞—è —á–∞—Å—Ç—å –±–æ–ª—å—à–µ 5 - –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ **–≤–≤–µ—Ä—Ö**
- –µ—Å–ª–∏ –¥—Ä–æ–±–Ω–∞—è —á–∞—Å—Ç—å –º–µ–Ω—å—à–µ 5 - –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ **–≤–Ω–∏–∑**

````python
print(a, '->', round(a))
print(b, '->', round(b))
print(c, '->', round(c))
````

üìå **–ü–æ–ª–µ–∑–Ω–æ –∑–Ω–∞—Ç—å**: –µ—Å–ª–∏ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π **—Ä–æ–≤–Ω–æ** `.5`, —Ç–æ –æ–∫—Ä—É–≥–ª—è–µ—Ç—Å—è –∫ –±–ª–∏–∂–∞–π—à–µ–º—É **—á–µ—Ç–Ω–æ–º—É** —á–∏—Å–ª—É:

````python
print(2.5, '->', round(2.5))  # –±—É–¥–µ—Ç 2 (—á–µ—Ç–Ω–æ–µ)
print(3.5, '->', round(3.5))  # –±—É–¥–µ—Ç 4 (—á–µ—Ç–Ω–æ–µ)
````

–ú–æ–∂–µ—Ç –æ–∫—Ä—É–≥–ª—è—Ç—å **–Ω–µ —Ç–æ–ª—å–∫–æ –¥–æ —Ü–µ–ª–æ–≥–æ**. –¢–æ–≥–¥–∞ –≤—Ç–æ—Ä—ã–º –∞—Ä–≥—É–º–µ–Ω—Ç–æ–º –Ω—É–∂–Ω–æ —É–∫–∞–∑–∞—Ç—å **—á–∏—Å–ª–æ –∑–Ω–∞–∫–æ–≤, –¥–æ –∫–æ—Ç–æ—Ä–æ–≥–æ —Ö–æ—Ç–∏–º –æ–∫—Ä—É–≥–ª–∏—Ç—å**:

````python
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –¥–æ 2—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
print(a, '->', round(a, 2))
````

````python
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –¥–æ 1-–≥–æ –∑–Ω–∞–∫–∞ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
print(a, '->', round(a, 1))
````

#### **`math.floor()`** - –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –≤–Ω–∏–∑ (2)
- –±–µ—Ä–µ–º –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ `math`
- –û—Ç–±—Ä–∞—Å—ã–≤–∞–µ—Ç –¥—Ä–æ–±–Ω—É—é —á–∞—Å—Ç—å, —Ç–æ –µ—Å—Ç—å –æ–∫—Ä—É–≥–ª—è–µ—Ç **–≤–Ω–∏–∑** –¥–æ –±–ª–∏–∂–∞–π—à–µ–≥–æ **—Ü–µ–ª–æ–≥–æ**

````python
# –í–∞—Ä–∏–∞–Ω—Ç 1
import math
print(a, '->', math.floor(a))
print(b, '->', math.floor(b))
print(c, '->', math.floor(c))
````

````python
# –í–∞—Ä–∏–∞–Ω—Ç 2
from math import *

print(a, '->', floor(a))
print(b, '->', floor(b))
print(c, '->', floor(c))
````

````python
# –í–∞—Ä–∏–∞–Ω—Ç 3 - –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å —Ç–æ–ª—å–∫–æ –Ω—É–∂–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é (–ø—Ä–æ–∑—Ä–∞—á–Ω–µ–µ + –º–µ–Ω—å—à–µ –ø–∞–º—è—Ç–∏ –∑–∞–Ω–∏–º–∞–µ—Ç)
from math import floor

print(a, '->', floor(a))
print(b, '->', floor(b))
print(c, '->', floor(c))
````

#### **`math.ceil()`** - –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –≤–≤–µ—Ä—Ö
- –±–µ—Ä–µ–º –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ `math`
- –û—Ç–±—Ä–∞—Å—ã–≤–∞–µ—Ç –¥—Ä–æ–±–Ω—É—é —á–∞—Å—Ç—å, —Ç–æ –µ—Å—Ç—å –æ–∫—Ä—É–≥–ª—è–µ—Ç **–≤–≤–µ—Ä—Ö** –¥–æ –±–ª–∏–∂–∞–π—à–µ–≥–æ **—Ü–µ–ª–æ–≥–æ**

````python
from math import ceil

print(a, '->', ceil(a))
print(b, '->', ceil(b))
print(c, '->', ceil(c))
````

## **3. –§–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ —Å—Ç—Ä–æ–∫**

<div style="background-color: rgba(56, 17, 90, 0.3); padding: 15px; margin: 15px; border-radius: 8px; border-left: 5px solid #8A2BE2;">

> **f-—Å—Ç—Ä–æ–∫–∞** ‚Äî —ç—Ç–æ –æ–±—ã—á–Ω–∞—è —Å—Ç—Ä–æ–∫–∞ (—Ç–µ–∫—Å—Ç –≤ –∫–∞–≤—ã—á–∫–∞—Ö), –Ω–æ –ø–µ—Ä–µ–¥ –Ω–µ–π —Å—Ç–∞–≤–∏–º –±—É–∫–≤—É `f`, –∞ –≤–Ω—É—Ç—Ä–∏ —Ñ–∏–≥—É—Ä–Ω—ã—Ö —Å–∫–æ–±–æ–∫ `{}` –º–æ–∂–Ω–æ –≤—Å—Ç–∞–≤–ª—è—Ç—å –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –∏–ª–∏ –¥–∞–∂–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è.
1. –ü–µ—Ä–µ–¥ —Å—Ç—Ä–æ–∫–æ–π —Å—Ç–∞–≤–∏–º `f` (–ª–∞—Ç–∏–Ω—Å–∫—É—é!). –ü—Ä–æ–±–µ–ª–∞ –º–µ–∂–¥—É `f` –∏ –æ—Ç–∫—Ä—ã–≤–∞—é—â–µ–π –∫–∞–≤—ã—á–∫–æ–π –±—ã—Ç—å **–Ω–µ** –¥–æ–ª–∂–Ω–æ.
2. –ï—Å–ª–∏ —Ö–æ—Ç–∏–º –≤—Å—Ç–∞–≤–∏—Ç—å –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é - –∑–∞–∫–ª—é—á–∞–µ–º –µ–µ –≤ —Ñ–∏–≥—É—Ä–Ω—ã–µ —Å–∫–æ–±–∫–∏ `{}`. 
</div>

**–ü—Ä–∏–º–µ—Ä**:

````python
name = "–ê–ª—å–±–µ—Ä—Ç"

# –ü—Ä–æ—Å—Ç–æ –ø–∏—à–µ–º f –ø–µ—Ä–µ–¥ –∫–∞–≤—ã—á–∫–æ–π. –ê –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é name –∑–∞–∫–ª—é—á–∞–µ–º –≤ —Ñ–∏–≥—É—Ä–Ω—ã–µ —Å–∫–æ–±–∫–∏
print(f"–ü—Ä–∏–≤–µ—Ç, {name}!")
````

–ê –≤–æ—Ç —Ç–∞–∫ –≤—ã–≥–ª—è–¥–µ–ª–æ –±—ã —Å –æ–±—ã—á–Ω—ã–º `print()`:

````python
print('–ü—Ä–∏–≤–µ—Ç, ' + name + '!')
````

‚ûï –ú–æ–∂–Ω–æ –≤—Å—Ç–∞–≤–ª—è—Ç—å –Ω–µ —Ç–æ–ª—å–∫–æ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ, –Ω–æ –∏ **–≤—ã—Ä–∞–∂–µ–Ω–∏—è**!

````python
a = 5
b = 3
print(f"{a} + {b} = {a + b}")
````

–ï—Å–ª–∏ —á–∏—Å–ª–æ —Å–ª–∏—à–∫–æ–º –¥–ª–∏–Ω–Ω–æ–µ (–º–Ω–æ–≥–æ –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π), –Ω–æ –º—ã —Ö–æ—Ç–∏–º:
1. –ï–≥–æ –∫—Ä–∞—Å–∏–≤–æ —Ä–∞—Å–ø–µ—á–∞—Ç–∞—Ç—å (—á—Ç–æ–±—ã –Ω–µ –∑–∞–Ω–∏–º–∞–ª–æ –≤–µ—Å—å –≤—ã–≤–æ–¥)
2. –°–æ—Ö—Ä–∞–Ω–∏—Ç—å –∏—Å—Ö–æ–¥–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π

–ú–æ–∂–Ω–æ –ø—Ä–∏–º–µ–Ω–∏—Ç—å –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ —Å—Ä–∞–∑—É –≤–Ω—É—Ç—Ä–∏ `print()`. –ù–∞–ø—Ä–∏–º–µ—Ä:

````python
pi = 3.14159265
print('—á–∏—Å–ª–æ –ø–∏ =', round(pi, 2))
````

–ù–æ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —á—É—Ç—å –ø—Ä–æ—â–µ - —Å –ø–æ–º–æ—â—å—é **f-—Å—Ç—Ä–æ–∫–∏**!

<div style="background-color: rgba(56, 17, 90, 0.3); padding: 15px; margin: 15px; border-radius: 8px; border-left: 5px solid #8A2BE2;">

**–û–∫—Ä—É–≥–ª–µ–Ω–∏–µ —Å –ø–æ–º–æ—â—å—é f-—Å—Ç—Ä–æ–∫–∏**

–ì–ª–∞–≤–Ω–æ–µ - –ø–æ—Å–ª–µ —á–∏—Å–ª–∞ –¥–æ–ª–∂–Ω—ã –Ω–∞–ø–∏—Å–∞—Ç—å –∞–±—Ä–∞–∫–∞–¥–∞–±—Ä—É –≤–∏–¥–∞ `:.Xf`, –≥–¥–µ `X` - —á–∏—Å–ª–æ, –¥–æ —Å–∫–æ–ª—å–∫–∏ –∑–Ω–∞–∫–æ–≤ —Ö–æ—Ç–∏–º –æ–∫—Ä—É–≥–ª–∏—Ç—å.

–ù–∞–ø—Ä–∏–º–µ—Ä:
- `:.1f` - –æ–∫—Ä—É–≥–ª–∏—Ç –¥–æ 1-–≥–æ –∑–Ω–∞–∫–∞ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
- `:.2f` - –æ–∫—Ä—É–≥–ª–∏—Ç –¥–æ 2-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
- –∏ —Ç.–¥.

</div>

````python
pi = 3.14159265

# –û–ë–Ø–ó–ê–¢–ï–õ–¨–ù–û - –ø–∏—à–µ–º f; —Ä–∞–±–æ—Ç–∞–µ—Ç —Ç–æ–ª—å–∫–æ –≤–Ω—É—Ç—Ä–∏ f-—Å—Ç—Ä–æ–∫–∏
print(f'—á–∏—Å–ª–æ –ø–∏ = {pi:.2f}')
````

üìå **–û–±—Ä–∞—Ç–∏ –≤–Ω–∏–º–∞–Ω–∏–µ**: –µ—Å–ª–∏ –∑–∞–±—ã–≤–∞–µ—à—å –ø—Ä–æ `f` –ø–µ—Ä–µ–¥ —Å—Ç—Ä–æ–∫–æ–π ‚Äî Python –Ω–µ –ø–æ–¥—Å—Ç–∞–≤–∏—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ!
