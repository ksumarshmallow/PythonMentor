–ú–æ–∂–Ω–æ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å —É—Å–ª–æ–≤–∏—è –∫–∞–∫ **–≤–æ–ø—Ä–æ—Å**, –Ω–∞ –∫–æ—Ç–æ—Ä—ã–π Python –æ—Ç–≤–µ—á–∞–µ—Ç "–¥–∞" –∏–ª–∏ "–Ω–µ—Ç" –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –µ–≥–æ –∏—Å—Ç–∏–Ω–Ω–æ—Å—Ç–∏.

- `True` ‚Äî –¥–∞, –ø—Ä–∞–≤–¥–∞

**‚ùóÔ∏è–í–ê–ñ–ù–û‚ùóÔ∏è** –ó–Ω–∞—á–µ–Ω–∏—è `True` –∏ `False`:

- –ë–µ–∑ –∫–∞–≤—ã—á–µ–∫! (–≠—Ç–æ –Ω–µ —Å—Ç—Ä–æ–∫–∏ ‚Äî —ç—Ç–æ –æ—Ç–¥–µ–ª—å–Ω—ã–π —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö!)

````python
# –ò —ç—Ç–æ –æ—Ç–¥–ª–µ–ª—å–Ω—ã–π —Ç–∏–ø - –ù–ï —Å—Ç—Ä–æ–∫–∏
print(type(True))
print(type(False))
````

### 1.2 –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ `bool` –≤ –¥—Ä—É–≥–∏–µ —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö

–ó–Ω–∞—á–µ–Ω–∏—è `bool` –º–æ–∂–Ω–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –≤ `int` (—Ü–µ–ª–æ–µ —á–∏—Å–ª–æ):
- `True` –ø—Ä–µ–≤—Ä–∞—â–∞–µ—Ç—Å—è –≤ `1`
- `False` –ø—Ä–µ–≤—Ä–∞—â–∞–µ—Ç—Å—è –≤ `0`

````python
print(int(True))   # 1
print(int(False))  # 0
````

–° `float` –±—É–¥–µ—Ç –ø–æ—Ö–æ–∂–∞—è –∏—Å—Ç–æ—Ä–∏—è:

````python
print(float(True))   # 1.0
print(float(False))  # 0.0
````

–ú–æ–∂–Ω–æ –∏ –Ω–∞–æ–±–æ—Ä–æ—Ç, –ø—Ä–µ–≤—Ä–∞—â–∞—Ç—å –¥—Ä—É–≥–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ `bool`:
- `0`, `0.0`, –ø—É—Å—Ç—ã–µ —Å—Ç—Ä–æ–∫–∏ ‚Äî —ç—Ç–æ `False`

## **2. –û–ø–µ—Ä–∞—Ç–æ—Ä—ã —Å—Ä–∞–≤–Ω–µ–Ω–∏—è**

–ö–æ–≥–¥–∞ –ø—Ä–æ–≥—Ä–∞–º–º–µ –Ω—É–∂–Ω–æ —Å—Ä–∞–≤–Ω–∏—Ç—å —á—Ç–æ-—Ç–æ, –º—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º **–æ–ø–µ—Ä–∞—Ç–æ—Ä—ã —Å—Ä–∞–≤–Ω–µ–Ω–∏—è**.
- –ë–æ–ª—å—à–µ –ª–∏ 5, —á–µ–º 3?
- –û–¥–∏–Ω–∞–∫–æ–≤—ã–µ –ª–∏ –∏–º–µ–Ω–∞ —É –¥–≤—É—Ö —É—á–µ–Ω–∏–∫–æ–≤?
- –î–ª–∏–Ω–Ω–µ–µ –ª–∏ —Å—Ç—Ä–æ–∫–∞ "–∫–æ—Ç" –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É, —á–µ–º "—Å–æ–±–∞–∫–∞"?

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, **–æ–ø–µ—Ä–∞—Ç–æ—Ä—ã —Å—Ä–∞–≤–Ω–µ–Ω–∏—è** *—Å—Ä–∞–≤–Ω–∏–≤–∞—é—Ç* –¥–≤–∞ –∑–Ω–∞—á–µ–Ω–∏—è –∏ –æ—Ç–≤–µ—á–∞—é—Ç:
- ‚úÖ `True` ‚Äî –µ—Å–ª–∏ —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ –≤–µ—Ä–Ω–æ
= ‚ùå `False` ‚Äî –µ—Å–ª–∏ –Ω–µ–≤–µ—Ä–Ω–æ

–†–µ–∑—É–ª—å—Ç–∞—Ç —Å—Ä–∞–≤–Ω–µ–Ω–∏—è ‚Äî —ç—Ç–æ —Ç–∏–ø **`bool`**: `True` –∏–ª–∏ `False`

<div style="background-color: rgba(56, 17, 90, 0.3); padding: 15px; margin: 15px; border-radius: 8px; border-left: 5px solid #8A2BE2;">

**–¢–∞–±–ª–∏—Ü–∞ –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–≤ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è**

| –ß—Ç–æ –ø—Ä–æ–≤–µ—Ä—è–µ–º    | –ö–∞–∫ –ø–∏—à–µ–º | –ü—Ä–∏–º–µ—Ä   | –†–µ–∑—É–ª—å—Ç–∞—Ç |
| ---------------- | --------- | -------- | --------- |
| –†–∞–≤–Ω–æ            | `==`      | `3 == 3` | `True`    |
| –ù–µ —Ä–∞–≤–Ω–æ         | `!=`      | `3 != 5` | `True`    |
| –ë–æ–ª—å—à–µ           | `>`       | `5 > 3`  | `True`    |
| –ú–µ–Ω—å—à–µ           | `<`       | `2 < 1`  | `False`   |
| –ë–æ–ª—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ | `>=`      | `5 >= 5` | `True`    |
| –ú–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ | `<=`      | `4 <= 6` | `True`    |

</div>

–° **—á–∏—Å–ª–∞–º–∏** –≤—Å–µ –ø—Ä–æ—Å—Ç–æ:

````python
print(3 > 2)     # True
print(5 == 10)   # False
print(8 <= 8)    # True
````

**–°—Ç—Ä–æ–∫–∏** —Ç–æ–∂–µ –º–æ–∂–Ω–æ —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å ‚Äî –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É, –∫–∞–∫ –≤ —Å–ª–æ–≤–∞—Ä–µ.

üìå **–ó–∞–ø–æ–º–∏–Ω–∞–µ–º**:
- Python —Å—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç —Å—Ç—Ä–æ–∫–∏ –ø–æ –±—É–∫–≤–∞–º —Å–ª–µ–≤–∞ –Ω–∞–ø—Ä–∞–≤–æ, –∏—Å–ø–æ–ª—å–∑—É—è –ø–æ—Ä—è–¥–∫–æ–≤—ã–π –Ω–æ–º–µ—Ä —Å–∏–º–≤–æ–ª–æ–≤ –≤ —Ç–∞–±–ª–∏—Ü–µ `Unicode`.
- –ó–∞–≥–ª–∞–≤–Ω—ã–µ –±—É–∫–≤—ã ‚â† —Å—Ç—Ä–æ—á–Ω—ã–µ. –ó–∞–≥–ª–∞–≤–Ω—ã–µ –±—É–∫–≤—ã **—Ä–∞–Ω—å—à–µ** –ø–æ –Ω–æ–º–µ—Ä—É ("–ê" < "–∞")

````python
print("–∫–æ—Ç" == "–∫–æ—Ç")     # True
print("–∫–æ—Ç" == "–ö–æ—Ç")     # False (—Ä–∞–∑–Ω—ã–π —Ä–µ–≥–∏—Å—Ç—Ä!)
print("–∞—Ä–±—É–∑" < "–±–∞–Ω–∞–Ω")  # True ("–∞" —Ä–∞–Ω—å—à–µ "–±")
print("–º–∞–º–∞" > "–ø–∞–ø–∞")    # False ("–º" —Ä–∞–Ω—å—à–µ "–ø")
````

**‚ùóÔ∏è–í–ê–ñ–ù–û‚ùóÔ∏è** –ú–æ–∂–Ω–æ —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å —Ç–æ–ª—å–∫–æ **–ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –æ–¥–Ω–æ–≥–æ —Ç–∏–ø–∞ –º–µ–∂–¥—É —Å–æ–±–æ–π**. –¢–æ –µ—Å—Ç—å —á–∏—Å–ª–∞ —Å —á–∏—Å–ª–∞–º–∏, —Å—Ç—Ä–æ–∫–∏ —Å–æ —Å—Ç—Ä–æ–∫–∞–º–∏ - —ç—Ç–æ –û–ö. –ê –≤–æ—Ç —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å —Å—Ç—Ä–æ–∫–∏ –∏ —á–∏—Å–ª–∞ ‚Äî —ç—Ç–æ –æ—à–∏–±–∫–∞!

````python
# –ù–µ–ª—å–∑—è —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å —Å—Ç—Ä–æ–∫—É –∏ —á–∏—Å–ª–æ - –±—É–¥–µ—Ç –æ—à–∏–±–∫–∞!
1 < "2"
````

>*–ò—Å–ø–æ–ª—å–∑—É—è –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã —Å—Ä–∞–≤–Ω–µ–Ω–∏—è, –º—ã —Å–æ–∑–¥–∞–µ–º **—É—Å–ª–æ–≤–∏–µ***

–ù–∞–ø—Ä–∏–º–µ—Ä, —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–≤—É—Ö —á–∏—Å–µ–ª - —ç—Ç–æ **—É—Å–ª–æ–≤–∏–µ**, —Ç–∞–∫ –∫–∞–∫ –º—ã —á—Ç–æ-—Ç–æ —Å–æ–ø–æ—Å—Ç–∞–≤–ª—è–µ–º –º–µ–∂–¥—É —Å–æ–±–æ–π, –ø–æ–ª—É—á–∞—è –Ω—É–∂–Ω—ã–π –¥–ª—è `if..elif` —Ç–∏–ø `bool`:

````python
# —ç—Ç–æ —É—Å–ª–æ–≤–∏–µ
3 > 7
````

````python
# —ç—Ç–æ —Ç–æ–∂–µ —É—Å–ª–æ–≤–∏–µ
1 != 2
````

````python
# –∏ —ç—Ç–æ - —É—Å–ª–æ–≤–∏–µ
"cat" == "dog"
````

**‚ùóÔ∏è–í–ê–ñ–ù–û‚ùóÔ∏è** –ü—Ä–∏ —Å—Ä–∞–≤–Ω–µ–Ω–∏–∏ —Å—Ç—Ä–æ–∫ –≤–∞–∂–µ–Ω –∏ **—Ä–µ–≥–∏—Å—Ç—Ä** –∏ **–ø—Ä–æ–±–µ–ª—ã** —Å **–ø—É–Ω–∫—Ç—É–∞—Ü–∏–µ–π**

````python
# —Å—Ç—Ä–æ–∫–∞ —Å –ø—Ä–æ–±–µ–ª–æ–º –∏ –±–µ–∑ –Ω–µ–µ - —Ä–∞–∑–Ω—ã–µ —Å—Ç—Ä–æ–∫–∏
"cat" == "cat "
````

````python
# —Å—Ç—Ä–æ–∫–∞ —Å –∑–∞–≥–ª–∞–≤–Ω–æ–π –∏ —Å—Ç—Ä–æ—á–Ω–æ–π –±—É–∫–≤—ã - —Ä–∞–∑–Ω—ã–µ —Å—Ç—Ä–æ–∫–∏
"cat" == "Cat"
````

````python
# –ø–æ –∞–Ω–∞–ª–æ–≥–∏–∏ —Å–ª–æ–≤–æ –∏ —Å–ª–æ–≤–æ —Å —Ç–æ—á–∫–æ–π - —Ä–∞–∑–Ω—ã–µ —Å—Ç—Ä–æ–∫–∏
"cat" == "cat."
````

## **3. –°–æ–∑–¥–∞–µ–º –ø—Ä–æ—Å—Ç—É—é —É—Å–ª–æ–≤–Ω—É—é –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏—é**

*–ö–∞–∫ –≤—Å–µ —ç—Ç–æ —Å–æ–µ–¥–∏–Ω—è–µ—Ç—Å—è?*
1. –ú—ã –ø–∏—à–µ–º —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ, —Å–æ–∑–¥–∞–≤–∞—è —É—Å–ª–æ–≤–∏–µ, –Ω–∞–ø—Ä–∏–º–µ—Ä `age >= 7`
2. Python –ø—Ä–æ–≤–µ—Ä—è–µ—Ç ‚Äî —ç—Ç–æ `True` –∏–ª–∏ `False`
    - –ï—Å–ª–∏ `True` ‚Äî —Å—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –±–ª–æ–∫ –ø–æ–¥ `if`
    - –ï—Å–ª–∏ `False` ‚Äî –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –±–ª–æ–∫ `else`

**–ü—Ä–∏–º–µ—Ä** - —Ö–æ—Ç–∏–º –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —á–µ—Ç–Ω–æ–µ –ª–∏ —á–∏—Å–ª–æ:
1. –ü–æ–ª—É—á–∞–µ–º —á–∏—Å–ª–æ `number`
2. –°–æ–∑–¥–∞–µ–º —É—Å–ª–æ–≤–∏–µ:
    - –∏—â–µ–º –æ—Å—Ç–∞—Ç–æ–∫ –¥–µ–ª–µ–Ω–∏—è —á–∏—Å–ª–∞ `number` –Ω–∞ 2 (`number % 2`)
    - –µ—Å–ª–∏ —ç—Ç–æ—Ç –æ—Å—Ç–∞—Ç–æ–∫ —Ä–∞–≤–Ω—è–µ—Ç—Å—è 0 $\to$ —á–∏—Å–ª–æ —á–µ—Ç–Ω–æ–µ
    - –µ—Å–ª–∏ —ç—Ç–æ—Ç –æ—Å—Ç–∞—Ç–æ–∫ –Ω–µ —Ä–∞–≤–Ω—è–µ—Ç—Å—è 0 $\to$ —á–∏—Å–ª–æ –Ω–µ—á–µ—Ç–Ω–æ–µ

–ó–¥–µ—Å—å **—É—Å–ª–æ–≤–∏–µ** - `number % 2 == 0`. –¢–∞–∫ –∫–∞–∫ —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è –Ω–∞ 2 (`number % 2`) —Å `0`.

````python
# –ø–æ–ª—É—á–∞–µ–º —á–∏—Å–ª–æ
number = 7

# –ø—Ä–æ–≤–µ—Ä—è–µ–º, —Ä–∞–≤–Ω—è–µ—Ç—Å—è –ª–∏ –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è –Ω–∞ 2 –Ω—É–ª—é
if number % 2 == 0:     # –Ω–µ –∑–∞–±—ã–≤–∞–µ–º –ø—Ä–æ –¥–≤–æ–µ—Ç–æ—á–∏–µ
    # –∏–Ω—Å—Ç—Ä—É–∫—Ü–∏—è –µ—Å–ª–∏ —á–∏—Å–ª–æ —á–µ—Ç–Ω–æ–µ
    # –Ω–µ –∑–∞–±—ã–≤–∞–µ–º –ø—Ä–æ –æ—Ç—Å—Ç—É–ø –≤ 4 –ø—Ä–æ–±–µ–ª–∞ ("–≤–Ω—É—Ç—Ä–∏" if)
    print("–ß–∏—Å–ª–æ —á–µ—Ç–Ω–æ–µ.")
else:                   # —Ç—É—Ç —Ç–æ–∂–µ –Ω—É–∂–Ω–æ –¥–≤–æ–µ—Ç–æ—á–∏–µ!
    # –±–ª–æ–∫ –∫–æ–¥–∞ –µ—Å–ª–∏ —á–∏—Å–ª–æ –Ω–µ—á–µ—Ç–Ω–æ–µ
    # –Ω–µ –∑–∞–±—ã–≤–∞–µ–º –ø—Ä–æ –æ—Ç—Å—Ç—É–ø –≤ 4 –ø—Ä–æ–±–µ–ª–∞ ("–≤–Ω—É—Ç—Ä–∏" if)
    print("–ß–∏—Å–ª–æ –Ω–µ—á–µ—Ç–Ω–æ–µ.")
````

–ú–æ–∂–Ω–æ –æ–ø—É—Å–∫–∞—Ç—å —á–∞—Å—Ç—å —Å `if`. –¢–∞–∫ –¥–µ–ª–∞—é—Ç, –µ—Å–ª–∏ –Ω–∞–º –Ω–∏—á–µ–≥–æ –Ω–µ –Ω—É–∂–Ω–æ –¥–µ–ª–∞—Ç—å, –∫–æ–≥–¥–∞ —É—Å–ª–æ–≤–∏–µ –Ω–µ–≤–µ—Ä–Ω–æ. –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ –º—ã —Ö–æ—Ç–∏–º —Ç–æ–ª—å–∫–æ –ø–µ—á–∞—Ç–∞—Ç—å —á–∏—Å–ª–æ, –µ—Å–ª–∏ –æ–Ω–æ —á–µ—Ç–Ω–æ–µ. –ù–æ –Ω–∏—á–µ–≥–æ –Ω–µ –¥–µ–ª–∞—Ç—å, –µ—Å–ª–∏ –æ–Ω–æ –Ω–µ—á–µ—Ç–Ω–æ–µ. –ú–æ–∂–µ—à—å –ø–æ–∏–≥—Ä–∞—Ç—å—Å—è —Å —á–∏—Å–ª–æ–º `number` –∏ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å, —á—Ç–æ –ø—Ä–æ—Å–∏—Ö–æ–¥–∏—Ç —Å –≤—ã–≤–æ–¥–æ–º –∫–æ–¥–∞.

````python
# –µ—Å–ª–∏ —Ç—É—Ç –Ω–µ—á–µ—Ç–Ω–æ–µ - –Ω–∏—á–µ–≥–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç—å –Ω–µ –±—É–¥–µ—Ç
# –∞ –µ—Å–ª–∏ —á–µ—Ç–Ω–æ–µ - –Ω–∞–ø–µ—á–∞—Ç–∞–µ—Ç—Å—è —Å–∞–º–æ —á–∏—Å–ª–æ
number = 7

if number % 2 == 0:
    print(number)
````
