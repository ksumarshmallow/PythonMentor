**–û—Å–Ω–æ–≤–Ω—ã–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏ Python**

| –û–ø–µ—Ä–∞—Ü–∏—è | –ù–∞–∑–≤–∞–Ω–∏–µ              | –ü—Ä–∏–º–µ—Ä   | –†–µ–∑—É–ª—å—Ç–∞—Ç |
| -------- | --------------------- | -------- | --------- |
| `+`      | —Å–ª–æ–∂–µ–Ω–∏–µ              | `5 + 2`  | `7`       |
| `-`      | –≤—ã—á–∏—Ç–∞–Ω–∏–µ             | `5 - 2`  | `3`       |
| `*`      | —É–º–Ω–æ–∂–µ–Ω–∏–µ             | `5 * 2`  | `10`      |
| `/`      | –¥–µ–ª–µ–Ω–∏–µ               | `5 / 2`  | `2.5`     |
| `//`     | —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–µ –¥–µ–ª–µ–Ω–∏–µ | `5 // 2` | `2`       |
| `%`      | –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è    | `5 % 2`  | `1`       |
| `**`     | –≤–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å  | `2 ** 3` | `8`       |

üí° –í—Å–µ —ç—Ç–∏ –æ–ø–µ—Ä–∞—Ü–∏–∏ –º–æ–∂–Ω–æ –¥–µ–ª–∞—Ç—å —Å –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏, —á–∏—Å–ª–∞–º–∏ –∏ –¥–∞–∂–µ —Å –≤—ã—Ä–∞–∂–µ–Ω–∏—è–º–∏ –≤–Ω—É—Ç—Ä–∏.




–ó–∞–¥–∞–¥–∏–º –¥–≤–∞ —á–∏—Å–ª–∞ `a` –∏ `b`:

````python
a = 100
b = 4
````

````python
# —Å–ª–æ–∂–µ–Ω–∏–µ: +
a + b
````

````python
# –≤—ã—á–∏—Ç–∞–Ω–∏–µ: -
a - b
````

````python
# —É–º–Ω–æ–∂–µ–Ω–∏–µ
a * b
````

````python
# –¥–µ–ª–µ–Ω–∏–µ - —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤—Å–µ–≥–¥–∞ float!
a / b
````

````python
# –¥–µ–ª–µ–Ω–∏–µ –Ω–∞—Ü–µ–ª–æ - —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤—Å–µ–≥–¥–∞ int
a // b
````

````python
# –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è
a % b
````

````python
# —Ç–æ–∂–µ –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç —Ü–µ–ª–æ–≥–æ –¥–µ–ª–µ–Ω–∏—è (15 % 4 = 15 - 4 * (15 // 4) = 15 - 4 * 3 = 15 - 12 = 3)
15 % 4
````

````python
# –≤–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å
2 ** 4
````

–û–±–æ–±—â–∏–º –≤ –æ–¥–Ω–æ–π —è—á–µ–π–∫–µ:
- –ó–∞–¥–∞–¥–∏–º –¥–≤–∞ —á–∏—Å–ª–∞ –∏ –ø–æ–ª–æ–∂–∏–º –∏—Ö –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ
- –í—ã–≤–µ–¥–µ–º –Ω–∞ —ç–∫—Ä–∞–Ω —ç—Ç–∏ –¥–≤–∞ —á–∏—Å–ª–∞
- –î–ª—è –∫–∞–∂–¥–æ–π –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–π –æ–ø–µ—Ä–∞—Ü–∏–∏ –±—É–¥–µ–º –ø–∏—Å–∞—Ç—å –µ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ —á–∏—Å–ª–∞–º–∏

````python
a = 10
b = 3

print("–ß–∏—Å–ª–∞:", a, b)
print("–°–ª–æ–∂–µ–Ω–∏–µ:", a + b)
print("–í—ã—á–∏—Ç–∞–Ω–∏–µ:", a - b)
print("–£–º–Ω–æ–∂–µ–Ω–∏–µ:", a * b)
print("–î–µ–ª–µ–Ω–∏–µ:", a / b)
print("–¶–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–µ –¥–µ–ª–µ–Ω–∏–µ:", a // b)
print("–û—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è:", a % b)
print("–í–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å:", a ** b)
````
